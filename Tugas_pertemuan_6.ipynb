```python
#buatlah output dari menggunakan bahasa pemrograman Python dengan :
#1 + 3 + 5 + 7 + 11 + 13 + 15 + 17 + 19 =. . .

jumlah = 0
string = ""

bilangan = 1
while bilangan <= 19:
  jumlah += bilangan
  string += str(bilangan)
  if bilangan < 19:
    string += '+'
  bilangan += 2
print(string, '=' ,jumlah )
```
